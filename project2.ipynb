```python
#using the links in the "link" column, download the first 1000 characters of each script
def get_script_text(url):
    try:
        if url.endswith('.pdf'):
            return "PDF file, skipped"
        
        response = requests.get(url, timeout=10)
        
        if 'text/html' in response.headers['Content-Type']:
            soup = BeautifulSoup(response.content, 'html.parser')
            return soup.get_text()[:1000]
        else:
            return "Non-HTML content, skipped"
    except Exception as e:
        return f"Error: {e}"

script_texts = scripts['link'].apply(get_script_text)
```

```python
#add a new column to the df with the text downloaded
scripts['script_text'] = script_texts
scripts.to_csv('pudding_texts.csv', index=False)
scripts.head()
```
